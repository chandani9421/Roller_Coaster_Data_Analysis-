```python
df= df[['coaster_name','Location','Status',
    'Manufacturer','year_introduced',
    'latitude', 'longitude','Type_Main',
    'opening_date_clean',
    'speed_mph','height_ft','Inversions','Gforce_clean']].copy()
```

```python
df = df.rename(columns ={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                    'height_ft':'Height_ft',
                    'Gforce_clean':'Gforce'
                    }) 
```

```python
df['Opening_Date'] = pd.to_datetime(df['Opening_Date'])
```

```python
df.isnull().sum()
```

```python
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

# feature understanding 

```python
df['Year_Introduced'].value_counts().head(10)
```

```python
sns.barplot(x='Year_Introduced', y='count', data=df['Year_Introduced'].value_counts().reset_index(name='count').head(10))
plt.title('Top 10 Years Coaster Introduced')
plt.xlabel('Year Introduced')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Speed_mph']
sns.histplot(df['Speed_mph'], bins=20, kde=True)
plt.title('Coaster Speeds (mph)')
plt.ylabel('Frequency')
```

```python
sns.scatterplot(y='Height_ft', x='Speed_mph',hue='Year_Introduced', data=df)
plt.grid()
plt.title('Coaster Height vs Speed')
```

```python
sns.pairplot(df, vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'], hue='Type_Main')
```

```python
df_correlation= df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_correlation
```

```python
# correlation heatmap
sns.heatmap(df_correlation, annot=True)
```

```python
# scatter plot with regression line

sns.lmplot(data=df ,x='Speed_mph', y= 'Height_ft')
```
